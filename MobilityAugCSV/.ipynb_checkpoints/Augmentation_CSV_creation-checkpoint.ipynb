## 3. Create augmented CSV files
- Creation of new CSV files using data augmentation on pre-existing data
- ONLY uses data w/ rare labels (Minimum Sway, Unlabeled, Sway rq UE sup)

```python
############
### Initialization
############

# ------DIRECTORIES------
normal_dir = r"C:\MobilityTemp\ScooterData" # Non-augmented dataset path
aug_dir = r"C:\MobilityTemp\AugData"   # Directory to save augmented dataset to
file_list = os.listdir(normal_dir)  # List of all CSV file names from normal directory

# ------EXTRA VARIABLES------
aug_funcs = [aug.jitter, aug.scaling, aug.magnitude_warp, aug.rotation]  # List of augmentation functions
aug_names = ["JIT_", "SCAL_", "MAG_", "ROT_"] # List of names of augmentation methods (for naming files)
col_names = "x0	y0	z0	x1	y1	z1	x2	y2	z2	x3	y3	z3	x4	y4	z4	x5	y5	z5	x6	y6	z6	x7	y7	z7	x8	y8	z8	label"
col_names = col_names.split('\t')  # List of all column names (to re-convert to dataframe)

time_step_sz = 16   # Size of each timestep
x_sz = len(col_names)  # Length of input matrix
```

```python
############
### Iteration
#############

# ---- Iterate through each augmentation method (outer loop) ----
for i in range(0, len(aug_names)):
    aug_matrix = np.empty((0, len(col_names)))   # Create empty array to prepare for appending

    # ---- Iterate through each csv file (inner loop) ----
    for path in file_list:

        # Create appropriate dataframes from CSV file
        df_norm = pd.read_csv(normal_dir + "\\" + path) # Dataframe from Nth csv file
        rare_df = df_norm.loc[df_norm["label"] != "Stable"]  # New dataframe of just data from rare labels
        if rare_df.size == 0:   # Break loop if a file contains NO rare labels
            continue

        # Augment numerical values, store in appropriate numpy array
        x_rare = rare_df.to_numpy()[:, 1:x_sz]    # Convert to Numpy arrays for use with the "utils" functions, and get all numerical values
        y_rare = rare_df.to_numpy()[:, x_sz]      # List of corresponding labels
        x_aug = aug_funcs[i](x_rare.reshape((-1, x_rare.shape[0], x_rare.shape[1]))) # Store "augmented" values using the ith augmentation method

        # Append the augmented data to the set
        for j in range(0, y_rare.size):
            # Add each row from the Nth file to the "final" matrix containing all rare data augmented with the appropriate augmentation method
            aug_matrix = np.concatenate((aug_matrix, [np.append(x_aug[0][j], y_rare[j])]))  # Create a matrix to convert back to a dataframe (in the same format as the original dataframe, but with the augmented values)
            
        # Make sure that the augmented data from the Nth file fits into the constraint of the time steps. Repeat the last row if nessecary
        if (y_rare.size % time_step_sz != 0):   # Check if the time step is a multiple of the size of the sample from the Nth file. If not, continue with the below steps.
            # NOTE: "Expected" value of the # of rows is a multiple of the time step, "actual" value is the original # of rows (for the Nth file)
            
            rem = y_rare.size % time_step_sz       # temporarily store the "remainder"

            rows_to_add = time_step_sz - rem       # Add the difference between the size of the time step and remainder (eg: for an "actual" value of 10 and a "expected" value of 9 with a time step of 3, since the remainder of 10/3 is 1, we need to add 1 row)
                
            last_row_indx = y_rare.size - 1        # get the last row's index
            
            for j in range(0, rows_to_add):
                # Add the final row of the Nth file to the "final" augmented matrix "rows_to_add" times
                aug_matrix = np.concatenate((aug_matrix, [np.append(x_aug[0][last_row_indx], y_rare[last_row_indx])]))
            
    
    # ----- END INNER LOOP -------
    
    
    # Save a file with appropriate data
    df_aug = pd.DataFrame(aug_matrix, columns=col_names)   # Format the matrix to a dataframe (this is now a dataframe containing labels + augmented data values)
    os.makedirs(aug_dir, exist_ok=True)     # Create
    df_aug.to_csv(aug_dir + "\\" + aug_names[i] + "RARE_LABELED.csv")    # In the "AugData" folder, create a CSV file of the augmented data with the corresponding augmentation method as a prefix (EG: using rotation, "ROT_RARE_LABELED.csv")
    
# ----- END OUTER LOOP -------
```

## #4. Additional notes about augmentation methods

- ### Used from: https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0254841
- ### General notes
    - The introduced augmentation methods are used for a variety of different types of data (image, signals, etc.), which may change the effectiveness of the augmentation methods.
    - More detail on actual augmentation methodology: https://github.com/terryum/Data-Augmentation-For-Wearable-Sensor-Data/blob/master/Example_DataAugmentation_TimeseriesData.ipynb
- ### Notes for Jittering
    - Formula: $ x'=x_1+\epsilon_1,...,x_t+\epsilon_t,...x_T+\epsilon_T $
    - $ x: $ some matrix with shape = **# samples (time)** by **# features**
    - $\epsilon:$ random noise from a gaussian distribution (that is pre-defined) added at each timestep t, uses: (https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)
- ### Notes for Magnitude Warping
    - Formula: $ x' = \alpha_1 x_1, ... ,\alpha_t x_t, ... ,\alpha_T x_T $
    - $ \alpha $: sequence created by interpolating a cubic spline with knots $ u = \upsilon_1, ... ,\upsilon_i, ... , \upsilon_I $
    - General spline: https://en.wikipedia.org/wiki/Spline_(mathematics)
    - Cubic spline: https://mathworld.wolfram.com/CubicSpline.html (?)
- ### Notes for Rotation
    - Formula: $ x' = Rx_1,...,R_xt,...,Rx_T $
    - $R$: element-wise random rotation matrix (either flipped or kept, for each element)
